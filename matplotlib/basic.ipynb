## Matplotlib example workflow

### 1. Prepare data

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
```

### 2. Set up plot

```python
fig, ax = plt.subplots(figsize=(10, 10))
```

### 3. Plot Data

```python
ax.plot(x, y)
```

### 4. Customize Plot

```python
ax.set_title("My Plot")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
```

### 5. Save & show (you save the whole figure)

```python
fig.savefig("../images/my-sample-plot.png")
```

## Making figures with Numpy arrays
- Line plot
- Scatter plot
- Bar plot
- Histogram

```python
# Create some data
x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data as line plot
plt.plot(x, x**2, label='power(x)')
```

```python
# Use the same data to plot a scatter plot
plt.scatter(x, np.sin(x), label='scatter(x)', color='red')
```

```python
# Make a plot from dictionary
nut_butter_prices = {
    'Almond Butter': 10.99,
    'Peanut Butter': 8.99,
    'Cashew Butter': 12.99
}
# make a bar plot
plt.bar(nut_butter_prices.keys(), nut_butter_prices.values())
plt.title("Nut Butter Prices")
plt.xlabel("Nut Butter Type")
plt.ylabel("Price ($)")
```

```python
# make a horizontal bar plot
plt.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
plt.title("Nut Butter")
plt.xlabel("Price ($)")
```

```python
# make a pie chart
plt.pie(nut_butter_prices.values(), labels=nut_butter_prices.keys(), autopct='%1.1f%%')
```

```python
# Make a histogram
data = np.random.randn(1000)
plt.hist(data, bins=30, edgecolor='black')
```

```python
# Make two subplots
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
# plot to each diferenct axis
ax1.plot(x, x/2, label='divide (x)')
ax2.scatter(np.random.random(10), np.random.random(10), color='green')
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values(), label='nut butter prices')
ax4.hist(np.random.randn(1000), bins=30, edgecolor='black', label='random data')
```

```python
# Option 2:
fig, ax = plt.subplots(2, 2, figsize=(10, 5))
# plot to each different axis
ax[0, 0].plot(x, x/2, label='divide (x)')
ax[0, 1].scatter(np.random.random(10), np.random.random(10), color='green')
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values(), label='nut butter prices')
ax[1, 1].hist(np.random.randn(1000), bins=30, edgecolor='black', label='random data')
# Show the plots
```

### Which one should you use? (pyplot vs matplotlib OO method)
* When plotting something quickly, okay to use puplot method
* When plotting something more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.target.values
```

```python
# Pyplot method to plot scatter plot
over_50.plot.scatter(x="age", y="chol", c="target", figsize=(10, 5), title="Cholesterol vs Age for Patients Over 50")
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax, title="Cholesterol vs Age for Patients Over 50")
# ax.set_xlim([45, 100])
ax.set_xlabel("Age")
ax.set_ylabel("Cholesterol")
```

## Customizing Matplotlib plots

```python
# See the different styles of plots available
plt.style.available
plt.style.use('tableau-colorblind10')
```

```python
car_sales.plot(x="Sale Date", y="Total Sales", kind="line", figsize=(10, 5), title="Total Sales Over Time")
```

```python
ax = df.plot(kind="bar", figsize=(10, 5), title="Random Data Bar Plot")
# Customize the plot with set() method
ax.set(title="Random Data Bar Plot from DataFrame", xlabel="Categories", ylabel="Values")
# Make the legend visible
ax.legend().set_visible(True)
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))
# Plot the data
scatter = ax.scatter(over_50["age"], over_50["chol"], c=over_50["target"], cmap="plasma") # this change the color map
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")
# Add a horizontal line
ax.axhline(over_50["chol"].mean(), color='blue', linewidth=0.8, linestyle='--')
# Add a color bar
cbar = fig.colorbar(scatter)
cbar.set_label("Target")
# Set the labels and title
ax.set_xlabel("Age")
ax.set_ylabel("Cholesterol")
ax.set_title("Cholesterol vs Age for Patients Over 50")
# Show the plot
plt.show()
```

```python
# Subplot of cholesterol, age, thalach
fig, (ax0, ax1) = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

# add data to ax0
scatter = ax0.scatter(over_50["age"], over_50["chol"], c=over_50["target"], cmap="winter")
# change the x-axis limits
ax0.set_xlim([50, 80])  # set x-axis limits
# customize ax0
ax0.set_title("Heart Disease - Cholesterol vs Age")
# ax0.set_xlabel("Age")
ax0.set_ylabel("Cholesterol")
# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# add a meanline
ax0.axhline(over_50["chol"].mean(), color='blue', linewidth=0.9, linestyle='--')

# add data to ax1
scatter = ax1.scatter(over_50["age"], over_50["thalach"], c=over_50["target"], cmap="autumn")
# change the x-axis limits
ax1.set_xlim([50, 80])  # set x-axis limits
ax1.set_ylim([65, 200])  # set y-axis limits
# customize ax1
ax1.set_title("Heart Disease - Max Heart Rate vs Age")
ax1.set_xlabel("Age")
ax1.set_ylabel("Max Heart Rate")
# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# add a meanline
ax1.axhline(over_50["thalach"].mean(), color='blue', linewidth=0.9, linestyle='--')

# Add title to the figure
fig.suptitle("Heart Disease Data for Patients Over 50", fontsize=16, fontweight='bold')
# Save the figure
fig.savefig("../images/heart-disease-over-50.png", dpi=300, bbox_inches='tight')
```
