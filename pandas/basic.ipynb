## Manipulate Data

```python
car_sales["Make"] = car_sales["Make"].str.lower()
car_sales["Colour"] = car_sales["Colour"].str.lower()
car_sales
```

```python
car_sales_md = pd.read_csv("../data/car-sales-missing-data.csv")
car_sales_md
```

```python
car_sales_md["Price"] = car_sales_md["Price"].str.replace('[/$/,/]', '', regex=True).astype(float)
car_sales_md["Make"] = car_sales_md["Make"].str.lower()
car_sales_md["Colour"] = car_sales_md["Colour"].str.lower()
car_sales_md
```

```python
car_sales_md["Odometer"] = car_sales_md["Odometer"].fillna(car_sales_md["Odometer"].mean())
car_sales_md
```

```python
# inplace=True means that the changes will be made to the original DataFrame
# car_sales_md_dropped_na = car_sales_md.dropna(inplace=True)
car_sales_md_dropped_na = car_sales_md.dropna()
car_sales_md
```

```python
car_sales_md_dropped_na
```

```python
# Columns from series
seats_column = pd.Series([5, 5, 5, 5, 5])

# New column called seats
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
car_sales
```

```python
# Column from Python list
fuel_economy = [7.5, 9.2, 5.0, 8.0, 9.3, 4.7, 6.5, 8.2, 7.0, 9.1]
car_sales["Fuel economy (L/100km)"] = fuel_economy
car_sales
```

```python
car_sales["Total fuel used"] = car_sales["Odometer (KM)"] / 100 * car_sales["Fuel economy (L/100km)"]
car_sales
```

```python
# Rename column Total fuel used to Total fuel used (L)
car_sales.rename(columns={"Total fuel used": "Total fuel used (L)"}, inplace=True)
car_sales
```

```python
# Create a column from a single value
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales["Passed road safety"] = True
car_sales.dtypes
```

```python
car_sales = car_sales.drop("Total fuel used (L)", axis=1)
car_sales
```
